** Now let us begin to focus on time information. What is the data type of the objects in the timeStamp column? **

```python
df['timeStamp'].dtype
```

** You should have seen that these timestamps are still strings. Use [pd.to_datetime](http://pandas.pydata.org/pandas-docs/stable/generated/pandas.to_datetime.html) to convert the column from strings to DateTime objects. **

```python
df['timeStamp'] = pd.to_datetime(df['timeStamp'])

df['timeStamp'].dtype
```

** You can now grab specific attributes from a Datetime object by calling them. For example:**

    time = df['timeStamp'].iloc[0]
    time.hour

**You can use Jupyter's tab method to explore the various attributes you can call. Now that the timestamp column are actually DateTime objects, use .apply() to create 3 new columns called Hour, Month, and Day of Week. You will create these columns based off of the timeStamp column, reference the solutions if you get stuck on this step.**

```python
# Retrieve hour attribute for first datetime element
df['timeStamp'].iloc[0].hour
```

```python
# Create 3 new columns called Hour, Month, and Day of Week from timeStamp column
df['Hour']=df['timeStamp'].dt.hour
df['Month']=df['timeStamp'].dt.month
df['Day_of_Week']=df['timeStamp'].dt.dayofweek
```

```python
df.head()
```

** Notice how the Day of Week is an integer 0-6. Use the .map() with this dictionary to map the actual string names to the day of the week: **

   

```python
# Map intger value of Day of Week column to string name using dictionary
dmap = {0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'}

df['Day_of_Week']=df['Day_of_Week'].apply(lambda x : dmap[x])

df.head()
```

** Now use seaborn to create a countplot of the Day of Week column with the hue based off of the Reason column. **

```python
# Examine the distribution of calls on each Day of Week with breakdown by Reason category
# Using countplot
# Using the order parameter to set the order of the x-axis labels
sns.countplot(
              x = 'Day_of_Week', 
              data = df, 
              hue= 'Reason', 
              order =('Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun' )
)
```

Trends:

Calls related to EMS are the most common, followed by Traffic and Fire.

The number of EMS calls fluctuate slightly from Mon to Sun but stay above 6000 calls per day.
Calls concerning Traffic fall during the weekends.
Fire calls are relatively stable throughout the week at around ~2000 calls per day

** Now create a gropuby object called byMonth, where you group the DataFrame by the month column and use the count() method for aggregation. Use the head() method on this returned DataFrame. **

```python
byMonth = df.groupby('Month').count().reset_index()

byMonth
```

** Now create a simple plot off of the dataframe indicating the count of calls per month. **

```python
byMonth.plot(x='Month', y='Reason')
plt.ylabel('Number of Calls')
plt.xlabel('Month')
```

THe lineplot shows that the hotline received the lowest number of calls From Aug to Dec, consistent with earlier observations.

Two other trends are clear here. The peak periods of 911 call are in Jan and July.

** Now see if you can use seaborn's lmplot() to create a linear fit on the number of calls per month. Keep in mind you may need to reset the index to a column. **

```python
sns.lmplot(x='Month', y='Reason', data=byMonth)
plt.xticks(range(14))
plt.show()

```

The regression plot showed the decline in number of calls from Jan to Dec. But the 95% CI spread is rather wide and we can also see that the points from Apr to July do not follow the linear trend.

**Create a new column called 'Date' that contains the date from the timeStamp column. You'll need to use apply along with the .date() method. ** 

```python
# Create a Date column by extracting date from the timeStamp column

df['Date']=df['timeStamp'].dt.date
```

** Now groupby this Date column with the count() aggregate and create a plot of counts of 911 calls.**

```python
byDate = df.groupby('Date').count().reset_index()

byDate
```

```python
byDate.plot(x='Date', y='Reason', legend=None)
plt.ylabel('Number of calls')
plt.xlabel('Date')
plt.title('Total Number of Calls Per Month')
```

** Now recreate this plot but create 3 separate plots with each plot representing a Reason for the 911 call**

```python
df_Fire = df[df['Reason']=='Fire']

byFire = df_Fire.groupby('Date').count().reset_index()

byFire.plot(x='Date', y='Reason', legend=None)
plt.ylabel('Number of calls')
plt.xlabel('Date')
plt.title('Number of Calls related to Fire Per Month')
```

```python
df_EMS = df[df['Reason']=='EMS']

byEMS = df_EMS.groupby('Date').count().reset_index()

byEMS.plot(x='Date', y='Reason', legend=None)
plt.ylabel('Number of calls')
plt.xlabel('Date')
plt.title('Number of Calls related to EMS Per Month')
```

```python
df_Traffic = df[df['Reason']=='Traffic']

byTraffic = df_Traffic.groupby('Date').count().reset_index()

byTraffic.plot(x='Date', y='Reason', legend=None)
plt.ylabel('Number of calls')
plt.xlabel('Date')
plt.title('Number of Calls related to Fire Per Month')
```

** Now let's move on to creating  heatmaps with seaborn and our data. We'll first need to restructure the dataframe so that the columns become the Hours and the Index becomes the Day of the Week. There are lots of ways to do this, but I would recommend trying to combine groupby with an [unstack](http://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.unstack.html) method. Reference the solutions if you get stuck on this!**

```python
df_heatmap = df.groupby(['Day_of_Week', 'Hour'])['Reason'].count().unstack()

df_heatmap
```

** Now create a HeatMap using this new DataFrame. **

```python
sns.heatmap(df_heatmap, cmap ="viridis")
```

** Now create a clustermap using this DataFrame. **

```python
sns.clustermap(df_heatmap, cmap ="viridis")
```

Busiest hours for the hotline are during the working week (Mon-Fri) afternoon, especially between 1600 and 1700 hours.

Lull period is between 2100 to 0600 hours throughout the week.
